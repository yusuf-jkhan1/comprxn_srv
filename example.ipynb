# Example of using Heroku FastAPI Service to implement k-means image compression

```python
import requests
from cmprxn.cmprxn import img_utils
```

```python
heroku_url = "https://cmprxn.herokuapp.com/cmprxn/cmprxn_kmeans"
```

```python
payload = {'path':'https://eventnook.s3.amazonaws.com/u/68110/coverimage_18042412170069_av-6643.jpg'}
```

```python
response = requests.post(url=heroku_url, params=payload)
```

```python
response
```

```python
json_response = response.json()
```

```python
json_response.keys()
```

```python
json_response['Centroids'][0:5]
```

```python
json_response['Labels'][0:5]
```

```python
img_utils_obj = img_utils('https://eventnook.s3.amazonaws.com/u/68110/coverimage_18042412170069_av-6643.jpg')
```

```python
img_utils_obj.display_image()
```

```python
img_utils_obj.reshape_img()
img_utils_obj.rebuild_compressed_image(centroids=json_response['Centroids'],labels=json_response['Labels'])
```
